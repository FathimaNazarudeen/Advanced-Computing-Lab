```python
import pandas as pd
data=pd.read_csv("/content/bank.csv")
```

```python
df=pd.DataFrame(data)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df=df.apply(le.fit_transform)
x=df[["job","marital","education","default","housing","loan","contact","month","poutcome"]]
y=df["deposit"]
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.9,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
ml = DecisionTreeClassifier(random_state=0,max_depth=2)
```

```python
model=ml.fit(x_train,y_train)
```

```python
res=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_1=accuracy_score(y_test,res)
print("Decision tree :",accuracy_1*100)
```

```python
from sklearn import tree
import graphviz
dot_data=tree.export_graphviz(ml,filled=True)
graph = graphviz.Source(dot_data)
graph
```
