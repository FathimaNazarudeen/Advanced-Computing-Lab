```python
import pandas as pd
from sklearn.datasets import load_iris
file=load_iris()
x=file.data
y=file.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
ml = DecisionTreeClassifier(random_state=0)
```

```python
model=ml.fit(x_train,y_train)
```

```python
res=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_1=accuracy_score(y_test,res)
print("Decision tree :",accuracy_1*100)
```

```python
from sklearn import tree
import graphviz
dot_data=tree.export_graphviz(ml)
graph = graphviz.Source(dot_data)
graph
```

```python
from sklearn.metrics import classification_report 
from sklearn.metrics import confusion_matrix
```

```python
import matplotlib.pyplot as plot
from sklearn.datasets import make_classification
from sklearn.metrics import plot_confusion_matrix
```

```python
plot_confusion_matrix(ml, x_test, y_test)  
plot.show()
```

```python
print("Classification Report for Decision Tree: \n", classification_report(y_test, res))
```

```python
from sklearn import metrics
```

```python
from sklearn import  metrics
fpr, tpr, thresholds = metrics.roc_curve(y_test,res,pos_label=2)
roc_auc = metrics.auc(fpr, tpr)
display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc,estimator_name='example estimator')
```

```python
import matplotlib.pyplot as plt
display.plot()
```
