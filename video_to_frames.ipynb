```python
# import required module
import glob
 
# get the path/directory
folder_dir = '/content/drive/MyDrive/ravdess dataset/angry'
 
# iterate over files in
# that directory
for images in glob.iglob(f'{folder_dir}/*'):
   
    # check if the image ends with png
    if (images.endswith(".mp4")):

        print(images)
        

        # Read the video from specified path
        cam = cv2.VideoCapture(images)


        # frame
        currentframe = 0

        while(True):
          
          # reading from frame
          ret,frame = cam.read()

          if ret:
            # if video is still left continue creating images
            name = './data/frame' + str(currentframe) + '.jpg'
            print ('Creating...' + name)

            # writing the extracted images
            cv2.imwrite(name, frame)

            # increasing counter so that it will
            # show how many frames are created
            currentframe += 1
          else:
            break

        # Release all space and windows once done
        cam.release()
        cv2.destroyAllWindows()


        plt.savefig("/content/drive/MyDrive/video_to_frames/angry/"+ images[45:-4])
        plt. clf()

        # # # AUDIO WITH NOISE
        # x = noise(x)
        # Xdb = librosa.amplitude_to_db(abs(X))
        # librosa.display.specshow(Xdb, sr=sr)
        # plt.savefig("/content/drive/MyDrive/audio_to_image/surprised/"+ images[45:-4]+" 02")
        # plt. clf()

        # # # STRETCHED AUDIO
        # x = stretch(x)
        # Xdb = librosa.amplitude_to_db(abs(X))
        # librosa.display.specshow(Xdb, sr=sr)
        # plt.savefig("/content/drive/MyDrive/audio_to_image/surprised/"+ images[45:-4]+" 03")
        # plt. clf()

        # # # AUDIO WITH PITCH
        # x = pitch(x, sr)
        # Xdb = librosa.amplitude_to_db(abs(X))
        # librosa.display.specshow(Xdb, sr=sr)
        # plt.savefig("/content/drive/MyDrive/audio_to_image/surprised/"+ images[45:-4]+" 04")
        # plt. clf()


```
