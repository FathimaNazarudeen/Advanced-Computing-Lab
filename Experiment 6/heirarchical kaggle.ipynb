```python
import pandas as pd
import numpy as np
```

```python
file=pd.read_csv("/content/heart.csv")
x=file[["age","sex","cp","trtbps","chol","fbs","restecg","thalachh","exng","oldpeak","slp","caa","thall"]][150:]
```

```python
from sklearn.decomposition import PCA
pcamodel=PCA(n_components=2)
principalcomponents=pcamodel.fit_transform(x)
newx=pd.DataFrame(data=principalcomponents,columns=["PC1","PC2"])
```

```python
xaxis=newx["PC1"]
yaxis=newx["PC2"]
newx
```

```python
from scipy.cluster import hierarchy
import matplotlib.pyplot as plt
Z = hierarchy.linkage(principalcomponents, 'complete')
plt.figure()
hierarchy.dendrogram(Z,orientation="top",above_threshold_color="red")
```
