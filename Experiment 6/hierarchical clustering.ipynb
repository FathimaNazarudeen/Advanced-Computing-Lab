```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_iris
file=load_iris()
X=file.data
Y=file.target
```

```python
from sklearn.decomposition import PCA
pcamodel=PCA(n_components=2)
principalcomponents=pcamodel.fit_transform(X)
```

```python
principalcomponents
```

```python
from scipy.cluster import hierarchy
import matplotlib.pyplot as plt
Z = hierarchy.linkage(principalcomponents, 'single')
plt.figure()
hierarchy.dendrogram(Z,orientation="top",above_threshold_color="red")
```
